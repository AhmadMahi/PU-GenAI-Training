## 1️⃣ OOP Foundations in AI Systems
We design AI systems as pipelines of components.

```
from abc import ABC, abstractmethod

class DataLoader(ABC):
    @abstractmethod
    def load_data(self):
        pass

class Preprocessor(ABC):
    @abstractmethod
    def process(self, data):
        pass

class Model(ABC):
    @abstractmethod
    def train(self, X, y):
        pass

    @abstractmethod
    def predict(self, X):
        pass
```

## 2️⃣ Concrete Implementations

```
import pandas as pd

class CSVDataLoader(DataLoader):
    def __init__(self, filepath):
        self.filepath = filepath

    def load_data(self):
        return pd.read_csv(self.filepath)
```

```
class AdvancedPreprocessor(Preprocessor):
    def process(self, data):
        data = data.dropna()
        data = data.select_dtypes(include=['number'])
        return data
```

```
from sklearn.linear_model import LogisticRegression

class LogisticModel(Model):
    def __init__(self):
        self.model = LogisticRegression()

    def train(self, X, y):
        self.model.fit(X, y)

    def predict(self, X):
        return self.model.predict(X)
```

## 3️⃣ Pipeline Architecture

```
class Pipeline:
    def __init__(self, loader, preprocessor, model):
        self.loader = loader
        self.preprocessor = preprocessor
        self.model = model

    def run(self):
        data = self.loader.load_data()
        data = self.preprocessor.process(data)

        X = data.iloc[:, :-1]
        y = data.iloc[:, -1]

        self.model.train(X, y)
        return self.model
```

## 4️⃣ Error Handling & Custom Exceptions

```
class AIException(Exception):
    pass

class DataLoadException(AIException):
    pass

class ModelTrainingException(AIException):
    pass
```

```
class SafeCSVLoader(DataLoader):
    def __init__(self, filepath):
        self.filepath = filepath

    def load_data(self):
        try:
            return pd.read_csv(self.filepath)
        except Exception as e:
            raise DataLoadException(f"Failed loading data: {e}")
```

## 5️⃣ Factory Pattern

```
class ModelFactory:
    @staticmethod
    def get_model(name):
        if name == "logistic":
            return LogisticModel()
        raise ValueError("Unknown model")
```

## 6️⃣ Strategy Pattern

```
class TrainingStrategy(ABC):
    @abstractmethod
    def train(self, model, X, y):
        pass

class BatchTraining(TrainingStrategy):
    def train(self, model, X, y):
        model.fit(X, y)
```

## 7️⃣ Builder Pattern

```
class PipelineBuilder:
    def __init__(self):
        self.loader = None
        self.preprocessor = None
        self.model = None

    def set_loader(self, loader):
        self.loader = loader
        return self

    def set_preprocessor(self, pre):
        self.preprocessor = pre
        return self

    def set_model(self, model):
        self.model = model
        return self

    def build(self):
        return Pipeline(self.loader, self.preprocessor, self.model)
```

## 8️⃣ Advanced: Ensemble Model

```
class EnsembleModel(Model):
    def __init__(self, models):
        self.models = models

    def train(self, X, y):
        for m in self.models:
            m.train(X, y)

    def predict(self, X):
        preds = [m.predict(X) for m in self.models]
        return sum(preds) / len(preds)
```
