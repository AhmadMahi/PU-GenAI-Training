# Smart Document Processing System (AI-Style, Pure Python)

```

import json
import time
from abc import ABC, abstractmethod
from datetime import datetime
```

```

class AIException(Exception):
    def __init__(self, message, component=None, severity="ERROR"):
        super().__init__(message)
        self.message = message
        self.component = component
        self.timestamp = datetime.now()
        self.severity = severity

    def __str__(self):
        return f"[{self.severity}] {self.timestamp} | {self.component} | {self.message}"

class InputException(AIException): pass
class ProcessingException(AIException): pass
```

```

class Logger:
    @staticmethod
    def log(level, message):
        print(f"[{level}] {datetime.now()} - {message}")
```

```

def run_with_safety(name, func, *args):
    try:
        Logger.log("INFO", f"Running {name}")
        return func(*args)
    except AIException as e:
        Logger.log("ERROR", str(e))
        return None
```

```

class InputHandler(ABC):
    @abstractmethod
    def load(self): pass

class Parser(ABC):
    @abstractmethod
    def parse(self, data): pass
```

```

class TextInputHandler(InputHandler):
    def __init__(self, text):
        self.text = text

    def load(self):
        if not self.text:
            raise InputException("Empty input", "InputHandler")
        return self.text
```

```

class SimpleParser(Parser):
    def parse(self, data):
        return data.lower().split()
```

```

class Pipeline:
    def __init__(self, input_handler, parser):
        self.input_handler = input_handler
        self.parser = parser

    def run(self):
        data = run_with_safety("Input", self.input_handler.load)
        tokens = run_with_safety("Parser", self.parser.parse, data)
        return tokens
```

```

pipeline = Pipeline(TextInputHandler("Test input"), SimpleParser())
print(pipeline.run())
```
